```python
new_df = df.dropna(axis=1) #drop NaN value columns
```

```python
new_df.head()
```

```python
plt.figure(figsize=(20,5), dpi=150)
sns.histplot(new_df, x='country_txt')
```

```python
#In order to get the readable and precise data we will sort according to number of terrorism occured in a country!
```

```python
group1 = new_df.groupby('country_txt').count()
```

```python
group1
```

```python
country_count = new_df['country_txt'].value_counts()
```

```python
#Occurence of terror events in the countries:
```

```python
country_count.head(15)
```

```python
#import new file 
```

```python
most_terror = pd.read_csv('file1.csv')
```

```python
most_terror.head()
```

```python
#Rename the columns
```

```python
most_terror.rename(columns={'Unnamed: 0':'Country','country_txt':'Terror_count'}, inplace=True)
```

```python
most_terror1 = most_terror.head(20)
```

```python
most_terror1
```

```python
plt.figure(figsize=(10,4), dpi=150)
plt.scatter(data=most_terror1, x='Country', y='Terror_count', marker='D')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of terrorist activity')
plt.title('Top 20 Terrorism hot zones');
```
